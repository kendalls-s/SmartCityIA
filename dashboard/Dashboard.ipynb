```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
# CNN (Basado en su classification_report)
labels_cnn = ['Normal', 'Sospechoso']
confusion_mtx = [[85, 15], [10, 90]]

#RNN (Basado en su variable 'future_predictions')
pasos_tiempo = np.arange(1, 11)
predicciones_rnn = [2.5, 2.7, 3.1, 2.8, 3.5, 4.0, 3.8, 4.2, 4.5, 4.3]
```

```python
# Dashboard
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Matriz de Confusión: Detección CCTV (CNN)",
        "Tendencia de Incidencias Futuras (RNN)",
        "Métricas de Precisión por Clase",
        "Distribución de Riesgo Proyectado"
    ),
    specs=[[{"type": "heatmap"}, {"type": "scatter"}],
           [{"type": "bar"}, {"type": "pie"}]]
)
```

```python
# Heatmap (CNN - Confusión)
fig.add_trace(
    go.Heatmap(z=confusion_mtx, x=labels_cnn, y=labels_cnn, colorscale='Viridis', showscale=False),
    row=1, col=1
)

# gráfico de línea (RNN - Predicciones)
fig.add_trace(
    go.Scatter(x=pasos_tiempo, y=predicciones_rnn, mode='lines+markers', name='Predicción RNN', line=dict(color='firebrick')),
    row=1, col=2
)

# Gráfico de barra (Métricas de Desempeño)
fig.add_trace(
    go.Bar(x=labels_cnn, y=[0.89, 0.92], name='Precisión', marker_color='skyblue'),
    row=2, col=1
)

# Grárfico de pastel (Distribución de Categorías)
fig.add_trace(
    go.Pie(labels=['Bajo Riesgo', 'Medio', 'Alto'], values=[40, 35, 25], hole=.3),
    row=2, col=2
)


fig.update_layout(
    title_text="Dashboard: Asistente Inteligente de Seguridad Urbana - Costa Rica",
    height=800,
    width=1000,
    showlegend=False,
    template="plotly_dark" # Estilo moderno para monitoreo
)

fig.show()
```
